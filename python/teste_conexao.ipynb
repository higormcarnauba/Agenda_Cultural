```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine, text

# 1. Carrega as variáveis do seu arquivo .env recém-criado
load_dotenv()

# 2. Monta a URL de conexão
db_url = f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"

# 3. Tenta conectar
try:
    engine = create_engine(db_url)
    with engine.connect() as connection:
        # Executa um comando simples SQL só para ver se responde
        result = connection.execute(text("SELECT current_database();"))
        print(f"✅ SUCESSO! Conectado ao banco: {result.fetchone()[0]}")
except Exception as e:
    print("❌ ERRO: Não foi possível conectar.")
    print("Verifique se a senha no .env está correta.")
    print(f"Detalhe do erro: {e}")
```
